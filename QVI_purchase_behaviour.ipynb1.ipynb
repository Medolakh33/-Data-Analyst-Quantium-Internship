DATA CHECKING AND CLEANING 

```python
# Check for Missing Data
missing_data_1 = purchase_behaviour.isnull().sum()
missing_data_2 = transaction.isnull().sum()
# Display  Missing Data
missing_data_1,missing_data_2
```

```python
#merged dataset
merged_data = pd.merge(purchase_behaviour, transaction, on='LYLTY_CARD_NBR', how='inner')

# Remove duplicate entries
merged_data.drop_duplicates(inplace=True)

# Display the merged dataset
print(merged_data)
merged_data
```

```python
# Define the current order of columns
current_columns_order = list(merged_data.columns)

# Define the desired order of columns
columns_order = ['DATE', 'LYLTY_CARD_NBR', 'STORE_NBR','PROD_NBR','TXN_ID','PREMIUM_CUSTOMER','LIFESTAGE','PROD_NAME','PROD_QTY','TOT_SALES']

# Reorder the three specific columns
qvi_data = merged_data.reindex(columns=columns_order)

# Display the organized DataFrame
print(qvi_data)
qvi_data
```

```python
#Checking merged data
qvi_data
```

Find outliers

```python

# Convert 'DATE' column to datetime format
qvi_data['DATE'] = pd.to_datetime(qvi_data['DATE'], origin='1899-12-30', unit='D')

# Calculate Z-score for 'TOT_SALES' column
z_scores = np.abs((qvi_data['TOT_SALES'] - qvi_data['TOT_SALES'].mean()) / qvi_data['TOT_SALES'].std())

# Define a threshold for outlier detection (Z-score greater than 3)
threshold = 3

# Find outliers
outliers = qvi_data[z_scores > threshold]

# Display outliers
print(outliers)
```

```python
# Analysis Monthly Total Sales
qvi_data['month'] = qvi_data['DATE'].dt.month
monthly_sales = qvi_data.groupby('month')['TOT_SALES'].sum()
plt.figure(figsize=(12, 5))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker='o')
plt.title('Monthly Total Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
# Understand customers' purchasing behavior
purchasing_behavior = qvi_data.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER']).agg(
    average_purchase_quantity=('PROD_QTY', 'mean'),
    total_sales=('TOT_SALES', 'sum'),
    most_frequent_product=('PROD_NAME', lambda x: x.mode()[0])  # Most frequently purchased product
)
purchasing_behavior.round(2)
```

```python
# Purchasing behavior for the Top 10 Most Sold Products

purchasing_behavior = qvi_data.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER']).agg(
    average_purchase_quantity=('PROD_QTY', 'mean'),
    total_sales=('TOT_SALES', 'sum'),
    most_frequent_product=('PROD_NAME', lambda x: x.mode()[0])  # Most frequently purchased product
)

purchasing_behavior.sort_values(by='total_sales', ascending=False, inplace=True)

# Display the sorted results
purchasing_behavior.head(10).round(2)
```

```python
# Average total sales per day
qvi_data['DATE'] = pd.to_datetime(qvi_data['DATE'])

total_sales_per_basket = qvi_data.groupby(['DATE','TXN_ID'])['TOT_SALES'].sum().reset_index()

average_sales_per_day = total_sales_per_basket.groupby('DATE')['TOT_SALES'].mean()
average_sales_per_day.head(3) , average_sales_per_day.tail(3)
```

```python
# Average Total Sales per Shopping Basket Over Time
qvi_data['DATE'] = pd.to_datetime(qvi_data['DATE'])

total_sales_per_basket = qvi_data.groupby(['DATE', 'TXN_ID'])['TOT_SALES'].sum().reset_index()

average_sales_per_day = total_sales_per_basket.groupby('DATE')['TOT_SALES'].mean().reset_index()

plt.figure(figsize=(12, 5))
plt.scatter(average_sales_per_day['DATE'], average_sales_per_day['TOT_SALES'], marker='o')
plt.title('Average Total Sales per Shopping Basket Over Time', fontsize=16)
plt.xlabel('Date', fontsize=10)
plt.ylabel('Average Total Sales per Shopping Basket', fontsize=14)
plt.grid(True)
plt.show()
```

```python
# Create a DataFrame
df = pd.DataFrame(qvi_data)

# Group by transaction ID and calculate the total quantity per transaction
total_quantity_per_transaction = df.groupby('TXN_ID')['PROD_QTY'].sum()

# Count multi-pack buyers (transactions with total quantity > 1)
multi_pack_buyers = (total_quantity_per_transaction > 1).value_counts()

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(multi_pack_buyers, labels=multi_pack_buyers.index, autopct='%1.1f%%', startangle=140)
plt.title(' Multi-Pack Buyers')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the plot
plt.show()
```

```python
# Total spending per customer
customer_spending = qvi_data.groupby('LYLTY_CARD_NBR')['TOT_SALES'].sum()

# Identify high-value customers (top 20% based on total spending)
high_value_customers = customer_spending[customer_spending >= customer_spending.quantile(0.8)]

# Total spending of all customers and high-value customers
total_spending = customer_spending.sum()
high_value_spending = high_value_customers.sum()

# Total spending contributed by high-value customers
percentage_high_value = (high_value_spending / total_spending) * 100
percentage_other = 100 - percentage_high_value

# Create labels for the pie chart
labels = ['High-Value Customers', 'Other Customers']
sizes = [percentage_high_value, percentage_other]
colors = ['skyblue', 'lightgray']
explode = (0.1, 0)  # explode the first slice (High-Value Customers)

plt.figure(figsize=(5, 5))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Proportion of Total Spending by High-Value Customers')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```
